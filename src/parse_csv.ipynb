```python
import pandas as pd
```

# Times and speedup

```python
df = pd.read_csv("../out/data.csv", delimiter=";")
df
```

```python
df["seconds"] = 0
```

```python
base_time = 353.5
```

```python
for i in range(11):
    time_string = df["time"][i]
    splitted_string = time_string.split(":")
    df["seconds"][i] = float(splitted_string[0]) * 60 + float(splitted_string[1])
```

```python
df
```

```python
df["speedup"] = base_time / df["seconds"]
df
```

```python
print("Max speedup = " + str(max(df["speedup"])))
print("Min time = " + str(min(df["seconds"]))+" seconds")
```

```python
df.to_csv("../out/processed_data.csv")
```
